```python
import matplotlib.pyplot as plt
times = ['Craig','Epoch','Gow','Grd','Pred','Sim','Train']
for i in times:
  f = open('/content/craig1/new_craig_2/'f'{i.lower()}''_time.txt','r')
  values=[]
  values_list=[]
  value_list=[]
  c=0
  avg_list=[]
  for values in f:
      values = values.split(' ')
      values[-1]=values[-1][:-1]
      values = list(map(float, values))
      values_list.append(values)
  epochs = list(range(0,len(values)))
  
  
  for i in range(0,len(values_list)):
    for j in range (0,len(values_list[0])):
      value_list.append(values_list[i][j])

  rows, cols = (10,5)
  idx_list = [[0 for i in range(cols)] for j in range(rows)]
  for i in range(0,10):
    c=i
    c1=0
    while c<50:
      idx_list[i][c1]=(value_list[c])
      c=c+10
      c1=c1+1
    avg_list.append(np.average(idx_list[i]))
  plt.plot(epochs,avg_list,color = 'g',label = f'{i}'' Time for Craig with 40% subset')
  
  plt.xlabel('Epoch', fontsize = 12)
  plt.ylabel('Time in Seconds', fontsize = 12)
    
```
